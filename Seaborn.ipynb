```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('countries-of-the-world.csv', index_col=0)
df
```

```python
gdp = df['GDP ($ per capita)']
gdp
phones = df['Phones (per 1000)']
phones
```

```python
# scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)
plt.show()
```

```python
percent_literate = df['Literacy (%)']
sns.scatterplot(x=gdp, y=percent_literate)
plt.show()
```

```python
region = df['Region']
# count plot with region on the y-axis
sns.countplot(y=region)

# Show plot
plt.show()
```

```python
mpg = pd.read_csv('mpg.csv', index_col = 0)
mpg
```

```python
# scatter plot of horsepower vs. mpg

sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            size="cylinders",hue='cylinders')

# Show plot
plt.show()
```

```python
sns.relplot(x='acceleration', y='mpg',data=mpg,kind='scatter',hue='origin',size='origin',style='origin')
plt.show()
```

```python
# line plot
sns.relplot(x='model_year',y='mpg',data=mpg,kind='line')
plt.show()
```

```python
#Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line",ci='sd')
plt.show()
```

```python
survey_data = pd.read_csv("young-people-survey-responses.csv",index_col=0)
survey_data
```

```python
sns.set_style('whitegrid')
sns.set_palette("RdBu")
# count plot of survey responses
# category_order = ["Never", "Rarely", "Sometimes", 
#                   "Often", "Always"]

sns.catplot(x="Parents' advice", 
            data=survey_data, 
            kind="count")
plt.show()
```

```python
sns.set_context('paper')

# Create bar plot
sns.catplot(x="Siblings", y="Loneliness",
            data=survey_data, kind="bar")
plt.show()
```

```python
sns.set_palette('Blues')

# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender",
                y="Age", data=survey_data, 
                kind="box", hue="Pets")

g.fig.suptitle('Age of Those Interested in Pets vs. Not')

plt.show()
```

```python
sns.set_style('dark')

# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Techno", 
                data=survey_data, kind="bar",
                col='Gender')

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
       ylabel="% Who Like Techno")
plt.show()
```
